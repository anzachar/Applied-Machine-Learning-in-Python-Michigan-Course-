### Question 2
What is the class distribution? (i.e. how many instances of `malignant` and how many `benign`?)

*This function should return a Series named `target` of length 2 with integer values and index =* `['malignant', 'benign']`

```python
def answer_two():
    
    # YOUR CODE HERE
    
    # creating a Series from the cancer.target array and counting the occurences of each class
    target=pd.Series(cancer.target).value_counts() 
    # mapping the indices of the Series to 'benign' and 'malignant'
    target.index=['benign', 'malignant']
    # Reordering the Series to ensure that 'malignant' is first 
    target=target.reindex(['malignant', 'benign'])
    
    return target
    
    
answer_two()
```

### Question 3
Split the DataFrame into `X` (the data) and `y` (the labels).

*This function should return a tuple of length 2:* `(X, y)`*, where* 
* `X` *has shape* `(569, 30)`
* `y` *has shape* `(569,)`.

```python
cancer_df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
    
# Add the target column to the DataFrame
cancer_df['target'] = cancer.target
def answer_three():
    # YOUR CODE HERE
    X=cancer_df.iloc[:, :-1]
    y=cancer_df.iloc[:,-1]
    return X,y
X,y=answer_three()
print(X.shape)
print(y.shape)
```

### Question 4
Using `train_test_split`, split `X` and `y` into training and test sets `(X_train, X_test, y_train, and y_test)`.

**Set the random number generator state to 0 using `random_state=0` to make sure your results match the autograder!**

*This function should return a tuple of length 4:* `(X_train, X_test, y_train, y_test)`*, where* 
* `X_train` *has shape* `(426, 30)`
* `X_test` *has shape* `(143, 30)`

```python
from sklearn.model_selection import train_test_split

def answer_four():
    # YOUR CODE HERE
    X, y = answer_three()  # Get X and y from answer_three
    X_train, X_test, y_train, y_test = train_test_split( X, y,  test_size=0.25,   # 25% of data for testing
    random_state=0    # Ensure reproducibility
    )
    return X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test=answer_four()
print(X_train.shape)  
print(X_test.shape)   
print(y_train.shape)  
print(y_test.shape)   
```

### Question 5
Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier with `X_train`, `y_train` and using one nearest neighbor (`n_neighbors = 1`).

*This function should return a `sklearn.neighbors.classification.KNeighborsClassifier`.

```python
from sklearn.neighbors import KNeighborsClassifier

def answer_five():
    # YOUR CODE HERE
    X_train, X_test, y_train, y_test=answer_four()
    knn=KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train, y_train)
    return knn
answer_five()    
```

### Question 6

### Optional plot

Try using the plotting function below to visualize the different predicition scores between train and test sets, as well as malignant and benign cells.

```python
def accuracy_plot():
    import matplotlib.pyplot as plt
    from sklearn.metrics import accuracy_score

    %matplotlib inline
    
    # YOUR CODE HERE
    knn=answer_five()
    X_train, X_test, y_train, y_test=answer_four()
    
    # Predicting labels for training and test sets
    y_train_pr=knn.predict(X_train)
    y_test_pr=knn.predict(X_test)
    
    # Calculating accuracy
    train_accuracy=accuracy_score(y_train, y_train_pr)
    test_accuracy=accuracy_score(y_test, y_test_pr)
    
    # Plotting
    labels=['Training Set', 'Test Set']
    accuracies=[train_accuracy, test_accuracy]
    
    # Creating bar plot
    plt.figure(figsize=(8, 6))
    plt.bar(labels, accuracies, color=['blue', 'magenta'])
    plt.xlabel('Dataset')
    plt.ylabel('Accuracy')
    plt.title('Accuracy Comparison: Training vs Test Set')
    plt.ylim(0, 1)  # Accuracy ranges from 0 to 1
    plt.show()
```
