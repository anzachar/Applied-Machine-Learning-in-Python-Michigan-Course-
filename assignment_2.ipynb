```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


np.random.seed(0)
n = 15
x = np.linspace(0,10,n) + np.random.randn(n)/5
y = np.sin(x)+x/6 + np.random.randn(n)/10


X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0)

def intro(): # improtant: delete 'notebook' and replace it with 'inline'
    %matplotlib inline 

    plt.figure()
    plt.scatter(X_train, y_train, label='training data')
    plt.scatter(X_test, y_test, label='test data')
    plt.legend(loc=4);

intro()
```

### Question 1

Write a function that fits a polynomial LinearRegression model on the *training data* `X_train` for degrees 1, 3, 6, and 9. (Use PolynomialFeatures in sklearn.preprocessing to create the polynomial features and then fit a linear regression model) For each model, find 100 predicted values over the interval x = 0 to 10 (e.g. `np.linspace(0,10,100)`) and store this in a numpy array. The first row of this array should correspond to the output from the model trained on degree 1, the second row degree 3, the third row degree 6, and the fourth row degree 9.

<img src="assets/polynomialreg1.png" style="width: 1000px;"/>

The figure above shows the fitted models plotted on top of the original data (using `plot_one()`).

<br>
*This function should return a numpy array with shape `(4, 100)`*

```python
def answer_one():
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import PolynomialFeatures
    
    degree_predictions = np.zeros((4,100))
    
    # YOUR CODE HERE
    
    # degrees
    deg=[1,3,6,9]
    
    # Array to store predictions
    pred=np.zeros((4,100))
    
    # Generating 100 points between 0 and 10
    x_pr=np.linspace(0,10,100).reshape(-1,1) # -1 ensures that the correct number of rows will be created
    
    # Iterating over each degree
    for i, deg in enumerate(deg):
        # Creating Polynomial features
        poly=PolynomialFeatures(degree=deg)
        X_train_poly=poly.fit_transform(X_train.reshape(-1,1))
        
        # Fitting linear regression model
        linreg=LinearRegression().fit(X_train_poly, y_train)
                                        
        # Transforming x_pr to polynomial features
        X_pr_poly=poly.transform(x_pr)
        
        # Predicting values for the x_pr array
        y_pr=linreg.predict(X_pr_poly)
        
        # Storing predictions in the correct row
        pred[i,:]=y_pr
    return pred
```

```python
# feel free to use the function plot_one() to replicate the figure 
# from the prompt once you have completed question one
def plot_one(degree_predictions):
    plt.figure(figsize=(10,5))
    plt.plot(X_train, y_train, 'o', label='training data', markersize=10)
    plt.plot(X_test, y_test, 'o', label='test data', markersize=10)
    for i,degree in enumerate([1,3,6,9]):
        plt.plot(np.linspace(0,10,100), degree_predictions[i], alpha=0.8, lw=2, label='degree={}'.format(degree))
    plt.ylim(-1,2.5)
    plt.legend(loc=4)

plot_one(answer_one())
```

### Question 2

Write a function that fits a polynomial LinearRegression model on the training data `X_train` for degrees 0 through 9. For each model compute the $R^2$ (coefficient of determination) regression score on the training data as well as the the test data, and return both of these arrays in a tuple.

*This function should return a tuple of numpy arrays `(r2_train, r2_test)`. Both arrays should have shape `(10,)`*

```python
def answer_two():
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.metrics import r2_score

    r2_train = np.array([])
    r2_test = np.array([])
    
    # YOUR CODE HERE
    for i in range(10):
        poly=PolynomialFeatures(degree=i)
        
        # Transforming the training data
        X_train_poly=poly.fit_transform(X_train.reshape(-1,1))
        
        # Fitting the linear regression model
        linreg=LinearRegression().fit(X_train_poly, y_train)
        
        # Computing R^2 score on training data
        train_score=linreg.score(X_train_poly, y_train)
        # Appending score to r2_train                                
        r2_train=np.append(r2_train, train_score)
        
        # Transforming the test data                                
        X_test_poly=poly.transform(X_test.reshape(-1,1)) 
        
        # Computing R^2 score on test data
        test_score=linreg.score(X_test_poly, y_test)
        r2_test=np.append(r2_test, test_score)
                                        
    return r2_train, r2_test
answer_two() 
```

### Question 5

Using `X_train` and `y_train` from the preceeding cell, train a DecisionTreeClassifier with default parameters and random_state=0. What are the 5 most important features found by the decision tree?

*This function should return a list of length 5 of the feature names in descending order of importance.*

```python
def answer_five():
    from sklearn.tree import DecisionTreeClassifier
    
    
    # YOUR CODE HERE
    
    # Training the Decision Tree Classifier
    tree_clf=DecisionTreeClassifier(random_state=0).fit(X_train2, y_train2)
    
    # Feature importances
    
    # Making a list of the feature importances
    importances=list(tree_clf.feature_importances_)
    # Sorting the list in descending order so the 5 most important features (larger values) are first
    importances_sorted=sorted(importances, reverse=True)
    # Making a new llist for the top 5 importances
    top5=importances_sorted[:5]
    
    # Finding the index of the top 5 importances in the original impoertances list  ## alternative: top5_ind = [importances.index(i) for i in top5] ##
    top5_ind=[]
    for importance in top5:
        index=importances.index(importance)
        top5_ind.append(index)
        
    # Matching the index to the name of the feature
    top5_name=[]
    importances_names=list(X_train2)
    for i in top5_ind:
        name=importances_names[i]
        top5_name.append(name)
    return top5_name


answer_five()

## ALTERNATIVE ##

#def answer_five():
    #from sklearn.tree import DecisionTreeClassifier
    #import numpy as np

    # Training the Decision Tree Classifier
    #tree_clf = DecisionTreeClassifier(random_state=0).fit(X_train2, y_train2)
    
    # Getting feature importances
    #importances = tree_clf.feature_importances_
    
    # Getting the indices of the top 5 most important features
    #top5_ind = np.argsort(importances)[-5:][::-1]  # argsort sorts indices, [-5:] gets the last 5, [::-1] reverses the order
    
    # Getting thr feature names corresponding to the top 5 indices
    #feature_names = X_train2.columns
    #top5_names = feature_names[top5_ind]
    
    #return top5_names.tolist()

# Getting the top 5 feature names
#top5_names = answer_five()
#print(top5_names)
```

### Question 6

For this question, use the `validation_curve` function in `sklearn.model_selection` to determine training and test scores for a Support Vector Classifier (`SVC`) with varying parameter values.

Create an `SVC` with default parameters (i.e. `kernel='rbf', C=1`) and `random_state=0`. Recall that the kernel width of the RBF kernel is controlled using the `gamma` parameter.  Explore the effect of `gamma` on classifier accuracy by using the `validation_curve` function to find the training and test scores for 6 values of `gamma` from `0.0001` to `10` (i.e. `np.logspace(-4,1,6)`).

For each level of `gamma`, `validation_curve` will use 3-fold cross validation (use `cv=3, n_jobs=2` as parameters for `validation_curve`), returning two 6x3 (6 levels of gamma x 3 fits per level) arrays of the scores for the training and test sets in each fold.

Find the mean score across the five models for each level of `gamma` for both arrays, creating two arrays of length 6, and return a tuple with the two arrays.

e.g.

if one of your array of scores is

    array([[ 0.5,  0.4,  0.6],
           [ 0.7,  0.8,  0.7],
           [ 0.9,  0.8,  0.8],
           [ 0.8,  0.7,  0.8],
           [ 0.7,  0.6,  0.6],
           [ 0.4,  0.6,  0.5]])
       
it should then become

    array([ 0.5,  0.73333333,  0.83333333,  0.76666667,  0.63333333, 0.5])

*This function should return a tuple of numpy arrays `(training_scores, test_scores)` where each array in the tuple has shape `(6,)`.*

```python
def answer_six():
    from sklearn.svm import SVC
    from sklearn.model_selection import validation_curve
    import numpy as np
    # YOUR CODE HERE
    
    # Creating list of gamma values:
    gamma_values=np.logspace(-4,1,6) # generates a sequence of numbers that are evenly spaced on a logarithmic scale (starting exponent:4, ending exponent:1, number of values: 6 
    
    # Training the SVC Classifier
    svc_clf=SVC(kernel='rbf', C=1, random_state=0)
    
    # Evaluating training and test scores
    training_scores_og, test_scores_og=validation_curve(svc_clf, X_mush, y_mush, param_name='gamma', param_range=gamma_values, cv=3, n_jobs=2) # the entire data set has to be used for some reason, though i believe this leads to leakage
    
    # Finding the mean for each level of gamma for training and test scores
    training_scores=np.mean(training_scores_og, axis=1) # Calculates the mean score across the folds (axis 1) for each value of gamma
    test_scores=np.mean(test_scores_og, axis=1)
    
    return training_scores, test_scores

answer_six() 
```
